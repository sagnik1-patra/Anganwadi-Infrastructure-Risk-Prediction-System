```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc
from sklearn.ensemble import RandomForestClassifier

# ===============================
# PATH
# ===============================

base_path = r"C:\Users\NXTWAVE\Downloads\Anganwadi Infrastructure Risk Prediction System"
dataset_path = os.path.join(base_path, "1_2.csv")
os.makedirs(base_path, exist_ok=True)

# ===============================
# LOAD & EXPAND DATA
# ===============================

df_original = pd.read_csv(dataset_path)

np.random.seed(42)
rows = 700

df = pd.DataFrame({
    "Block": np.random.choice(df_original["Block"], rows),
    "Rainfall_mm": np.random.normal(1000, 200, rows),
    "Groundwater_Level_m": np.random.normal(10, 3, rows),
    "Population_Growth_%": np.random.uniform(1, 5, rows),
    "Water_Scarcity_Index": np.random.uniform(0, 1, rows)
})

df["Risk"] = (
    (df["Rainfall_mm"] < 850) &
    (df["Groundwater_Level_m"] > 11)
).astype(int)

features = [
    "Rainfall_mm",
    "Groundwater_Level_m",
    "Population_Growth_%",
    "Water_Scarcity_Index"
]

X = df[features]
y = df["Risk"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# =====================================================
# BAT ALGORITHM (BA) OPTIMIZATION
# =====================================================

def BA_optimize():
    best_score = 0
    best_params = None
    
    for _ in range(25):  # bat iterations
        n_estimators = int(np.random.uniform(100, 500))
        max_depth = int(np.random.uniform(3, 15))
        
        model = RandomForestClassifier(
            n_estimators=n_estimators,
            max_depth=max_depth,
            random_state=42
        )
        
        model.fit(X_train, y_train)
        acc = accuracy_score(y_test, model.predict(X_test))
        
        if acc > best_score:
            best_score = acc
            best_params = (n_estimators, max_depth)
    
    return best_params

best_params = BA_optimize()

# =====================================================
# AIS OPTIMIZATION
# =====================================================

def AIS_optimize():
    best_score = 0
    best_weights = None
    
    for _ in range(25):
        weights = np.random.uniform(0.5, 1.5, len(features))
        
        model = RandomForestClassifier(
            n_estimators=best_params[0],
            max_depth=best_params[1],
            random_state=42
        )
        
        model.fit(X_train * weights, y_train)
        acc = accuracy_score(y_test, model.predict(X_test * weights))
        
        if acc > best_score:
            best_score = acc
            best_weights = weights
    
    return best_weights

best_weights = AIS_optimize()

# =====================================================
# FINAL HYBRID MODEL
# =====================================================

bis_model = RandomForestClassifier(
    n_estimators=best_params[0],
    max_depth=best_params[1],
    random_state=42
)

bis_model.fit(X_train * best_weights, y_train)

bis_preds = bis_model.predict(X_test * best_weights)
bis_probs = bis_model.predict_proba(X_test * best_weights)[:, 1]
bis_accuracy = accuracy_score(y_test, bis_preds)

# =====================================================
# SAVE MODEL
# =====================================================

joblib.dump(bis_model, os.path.join(base_path, "bis_model.pkl"))

# =====================================================
# SAVE RESULTS
# =====================================================

df["bis_Risk_Prediction"] = bis_model.predict(X * best_weights)
df["bis_Risk_Probability"] = bis_model.predict_proba(X * best_weights)[:, 1]

df.to_csv(os.path.join(base_path, "bis_results.csv"), index=False)

prediction_json = df.head(20)[["Block", "bis_Risk_Prediction", "bis_Risk_Probability"]].to_dict(orient="records")

with open(os.path.join(base_path, "bis_predictions.json"), "w") as f:
    json.dump(prediction_json, f, indent=4)

# Save config
config = {
    "model": "Hybrid_BA_AIS_RandomForest",
    "accuracy": float(bis_accuracy),
    "best_n_estimators": best_params[0],
    "best_max_depth": best_params[1],
    "feature_weights": best_weights.tolist()
}

with open(os.path.join(base_path, "bis_config.yaml"), "w") as file:
    yaml.dump(config, file)

# =====================================================
# GRAPHS (SHOW + SAVE)
# =====================================================

# Confusion Matrix
cm = confusion_matrix(y_test, bis_preds)
plt.figure()
sns.heatmap(cm, annot=True, fmt="d")
plt.title("BIS Confusion Matrix")
plt.savefig(os.path.join(base_path, "bis_confusion_matrix.png"))
plt.show()

# Accuracy Graph
plt.figure()
plt.bar(["BA + AIS Hybrid"], [bis_accuracy])
plt.title("BIS Model Accuracy")
plt.savefig(os.path.join(base_path, "bis_accuracy_graph.png"))
plt.show()

# Feature Importance
plt.figure()
plt.bar(features, bis_model.feature_importances_)
plt.xticks(rotation=45)
plt.title("BIS Feature Importance")
plt.tight_layout()
plt.savefig(os.path.join(base_path, "bis_feature_importance.png"))
plt.show()

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, bis_probs)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
plt.plot([0, 1], [0, 1], linestyle='--')
plt.legend()
plt.title("BIS ROC Curve")
plt.savefig(os.path.join(base_path, "bis_roc_curve.png"))
plt.show()

print("\n BIS HYBRID MODEL COMPLETED")
print("Accuracy:", bis_accuracy)
```
