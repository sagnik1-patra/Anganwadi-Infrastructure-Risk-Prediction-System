```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc
from sklearn.ensemble import RandomForestClassifier

# ===============================
# PATH
# ===============================

base_path = r"C:\Users\NXTWAVE\Downloads\Anganwadi Infrastructure Risk Prediction System"
dataset_path = os.path.join(base_path, "1_2.csv")

os.makedirs(base_path, exist_ok=True)

# ===============================
# LOAD DATA
# ===============================

df_original = pd.read_csv(dataset_path)

np.random.seed(42)
rows = 500

df = pd.DataFrame({
    "Block": np.random.choice(df_original["Block"], rows),
    "Rainfall_mm": np.random.normal(1000, 200, rows),
    "Groundwater_Level_m": np.random.normal(10, 3, rows),
    "Population_Growth_%": np.random.uniform(1, 5, rows),
    "Water_Scarcity_Index": np.random.uniform(0, 1, rows)
})

df["Risk"] = (
    (df["Rainfall_mm"] < 850) &
    (df["Groundwater_Level_m"] > 11)
).astype(int)

features = [
    "Rainfall_mm",
    "Groundwater_Level_m",
    "Population_Growth_%",
    "Water_Scarcity_Index"
]

X = df[features]
y = df["Risk"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# ====================================================
# CSA OPTIMIZATION
# ====================================================

def CSA_optimize():
    best_score = 0
    best_params = None
    for _ in range(15):
        n_estimators = np.random.randint(100, 400)
        max_depth = np.random.randint(3, 10)
        model = RandomForestClassifier(
            n_estimators=n_estimators,
            max_depth=max_depth,
            random_state=42
        )
        model.fit(X_train, y_train)
        acc = accuracy_score(y_test, model.predict(X_test))
        if acc > best_score:
            best_score = acc
            best_params = (n_estimators, max_depth)
    return best_params

best_params = CSA_optimize()

# ====================================================
# AIS OPTIMIZATION
# ====================================================

def AIS_optimize():
    best_score = 0
    best_weights = None
    for _ in range(15):
        weights = np.random.uniform(0.5, 1.5, len(features))
        model = RandomForestClassifier(
            n_estimators=best_params[0],
            max_depth=best_params[1],
            random_state=42
        )
        model.fit(X_train * weights, y_train)
        acc = accuracy_score(y_test, model.predict(X_test * weights))
        if acc > best_score:
            best_score = acc
            best_weights = weights
    return best_weights

best_weights = AIS_optimize()

# ====================================================
# FINAL HYBRID MODEL
# ====================================================

hybrid_model = RandomForestClassifier(
    n_estimators=best_params[0],
    max_depth=best_params[1],
    random_state=42
)

hybrid_model.fit(X_train * best_weights, y_train)

hybrid_preds = hybrid_model.predict(X_test * best_weights)
hybrid_probs = hybrid_model.predict_proba(X_test * best_weights)[:, 1]
hybrid_accuracy = accuracy_score(y_test, hybrid_preds)

# ====================================================
# SAVE MODEL
# ====================================================

joblib.dump(hybrid_model, os.path.join(base_path, "hybrid_model.pkl"))

# ====================================================
# SAVE CSV + JSON
# ====================================================

df["Hybrid_Risk_Prediction"] = hybrid_model.predict(X * best_weights)
df["Hybrid_Risk_Probability"] = hybrid_model.predict_proba(X * best_weights)[:, 1]

df.to_csv(os.path.join(base_path, "hybrid_results.csv"), index=False)

prediction_json = df.head(20)[["Block", "Hybrid_Risk_Prediction", "Hybrid_Risk_Probability"]].to_dict(orient="records")

with open(os.path.join(base_path, "hybrid_predictions.json"), "w") as f:
    json.dump(prediction_json, f, indent=4)

# ====================================================
# GRAPHS (SHOW + SAVE)
# ====================================================

# 1 Confusion Matrix
cm = confusion_matrix(y_test, hybrid_preds)
plt.figure()
sns.heatmap(cm, annot=True, fmt="d")
plt.title("Hybrid Confusion Matrix")
plt.savefig(os.path.join(base_path, "hybrid_confusion_matrix.png"))
plt.show()

# 2 Accuracy Graph
plt.figure()
plt.bar(["Hybrid AIS+CSA"], [hybrid_accuracy])
plt.title("Hybrid Model Accuracy")
plt.savefig(os.path.join(base_path, "hybrid_accuracy_graph.png"))
plt.show()

# 3 Feature Importance
importances = hybrid_model.feature_importances_
plt.figure()
plt.bar(features, importances)
plt.xticks(rotation=45)
plt.title("Hybrid Feature Importance")
plt.tight_layout()
plt.savefig(os.path.join(base_path, "hybrid_feature_importance.png"))
plt.show()

# 4 ROC Curve
fpr, tpr, _ = roc_curve(y_test, hybrid_probs)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
plt.plot([0, 1], [0, 1], linestyle='--')
plt.title("Hybrid ROC Curve")
plt.legend()
plt.savefig(os.path.join(base_path, "hybrid_roc_curve.png"))
plt.show()

# 5 Risk Distribution
plt.figure()
sns.countplot(x=df["Hybrid_Risk_Prediction"])
plt.title("Risk Prediction Distribution")
plt.savefig(os.path.join(base_path, "hybrid_risk_distribution.png"))
plt.show()

# 6 Probability Histogram
plt.figure()
plt.hist(df["Hybrid_Risk_Probability"], bins=20)
plt.title("Risk Probability Distribution")
plt.savefig(os.path.join(base_path, "hybrid_probability_histogram.png"))
plt.show()

# 7 Correlation Heatmap
plt.figure()
sns.heatmap(df[features + ["Risk"]].corr(), annot=True)
plt.title("Feature Correlation Heatmap")
plt.savefig(os.path.join(base_path, "hybrid_correlation_heatmap.png"))
plt.show()

print("\n HYBRID MODEL COMPLETED")
print("Hybrid Accuracy:", hybrid_accuracy)
```
